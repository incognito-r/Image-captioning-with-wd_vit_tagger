```python
import pandas as pd
```

```python
mapping_file_path = "data/CelebAMask-HQ/CelebA-HQ-Identity.csv"

# load csv
id_map = pd.read_csv(mapping_file_path)
id_map.head(10)
```

```python
# import csv

high_res_path = 'data/CelebA-RAW/CelebA-HIGH-RES.csv'
mid_res_path = 'data/CelebA-RAW/CelebA-MID-RES.csv'
low_res_path = 'data/CelebA-RAW/CelebA-LOW-RES.csv'


high_res_df = pd.read_csv(high_res_path)
mid_res_df = pd.read_csv(mid_res_path)
low_res_df = pd.read_csv(low_res_path)
```

```python
high_res_df
```

```python
import pandas as pd

subset_df = id_map[['image_id', 'orig_file']].copy()
subset_df['orig_id'] = subset_df['orig_file'].str.replace('.jpg', '', case=False)

# Extract image ID from 'orig_file' and convert to int16
subset_df['orig_id'] = subset_df['orig_id'].astype('Int32')

# Step 3: Create sets for fast lookup
high_ids = set(high_res_df['image_id'])
mid_ids = set(mid_res_df['image_id'])
low_ids = set(low_res_df['image_id'])

# Step 4: Define reference lookup function
def get_reference_code(image_id):
    if image_id in high_ids:
        return 0
    elif image_id in mid_ids:
        return 1
    elif image_id in low_ids:
        return 2
    else:
        return -1  # Not found

# Step 5: Apply function
subset_df['high0_mid1_low2'] = subset_df['orig_id'].apply(get_reference_code)

# Optional: preview result
subset_df.head(10)
```

```python
# save dataframe
subset_df.to_csv('data/CelebAMask-HQ/CelebA-HQ-Identity.csv', index=False)
```
